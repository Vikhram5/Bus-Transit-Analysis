```python
import pandas as pd
df=pd.read_csv('04-02-19.csv')
```

```python
cols=['Schedule Name','Adult','Ticket Issued Time','From Stage','To Stage','Source','Destination']
df=df[cols]
bus_no =519
df = df[df['Schedule Name'].str.contains(rf'\b{bus_no}\b')]
```

```python
#df=df[(df['Source']=='T.NAGAR') & (df['Destination']=='THIRUPORUR')]
df=df[(df['From Stage']=='ANNA UNIV')]
df.sort_values('Ticket Issued Time')
#print(df.to_string())
```

```python
df.sort_values('Ticket Issued Time')
```

```python
df['Ticket Issued Time'] = pd.to_datetime(df['Ticket Issued Time'], format='%H:%M:%S', errors='coerce')
df['Hour'] = df['Ticket Issued Time'].dt.hour
df['Minute'] = df['Ticket Issued Time'].dt.minute
df=df[df['Hour']==9]
print(df.to_string())
```

```python
print("ANNA UNIV\n")
min_times = df.groupby('Schedule Name').agg({'Hour': 'min', 'Minute': 'min'}).reset_index()

# Display the result
print(min_times)
```

```python
average_hour = df['Hour'].mean()
average_minute = df['Minute'].mean()
average_minute_rounded = round(average_minute)
print(round(average_minute))
```

```python
import datetime
average_time = datetime.time(int(average_hour), int(average_minute_rounded))
print("Average Time:", average_time)
#ANNA UNIV
```
